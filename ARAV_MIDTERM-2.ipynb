**Part 0: Importing**

```python
import pandas as pd
import numpy as np
import plotnine
from plotnine import *
%matplotlib inline
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQ39MFe0DOqXzb15kZa_6aTYFbqvxd7744VZ2N_qx5DcGynjwEz4ndle29a-s8jwwkk6LNRHhcT99VK/pub?gid=907846205&single=true&output=csv"
```

```python
df = pd.read_csv(url) 
df.head()
```

**Part A: Cleaning Up**

```python
#Step 1 - Removing some unneccessary columns to see the ones that have been hidden

cols = [1,6,9]
df.drop(df.columns[cols],axis=1,inplace=True)
df.head()
```

```python
#Step 2 - Rename useful columns for better aesthetics

dict = {'Yearly brutto salary (without bonus and stocks) in EUR': 'Base Salary',
        'Yearly bonus + stocks in EUR' : 'Bonus + Stocks',
        'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country' : 
        'Last Year Base Salary',
        'Annual bonus+stocks one year ago. Only answer if staying in same country' : 'Last Year Bonus + Stocks',
       'Your main technology / programming language' : 'Programming Language',
       'Total years of experience' : 'Experience'}



df.rename(columns=dict,inplace=True)

df.head()
```

```python
#Step 3: Extract all useful columns into new df

df_refined = df[['Gender', 'Experience', 'Seniority level', 'Base Salary', 'Bonus + Stocks', 'Last Year Base Salary', 'Programming Language', 
                 'Last Year Bonus + Stocks', 'Employment status']]
df_refined.head()
```

**Part B: Refining**

```python
#Step 1: Replace NaN Values with 0 to enable comparison in total compensation (what if NaN != 0 tho)

df_refined = df_refined.fillna(0)

# Let's Capitalize the Programming languages so we can properly count them. 


df['Programming Language'] = df['Programming Language'].str.upper()


```

```python
#Step 2: Create new columns for total comp (base + bonus + stocks) for this year and last year


#Part i: Convert the column 'Bonus + Stocks' into a type that allows for operations to be carried out - sort of a brute-force method here but it's effective
df_refined['Bonus + Stocks'] = df_refined['Bonus + Stocks'].astype('str')
df_refined['Bonus + Stocks'] = df_refined['Bonus + Stocks'].str.replace(',', '')
df_refined['Bonus + Stocks'] = pd.to_numeric(df_refined['Bonus + Stocks'], errors='coerce')


#Part ii: Add Base + Bonus + Stocks to get total comp, and put this in a new column
df_refined['Total Comp'] = df_refined['Bonus + Stocks'] + df_refined['Base Salary']

#Part iii: Repeat for last year's data
df_refined['Last Year Bonus + Stocks'] = df_refined['Last Year Bonus + Stocks'].astype('str')
df_refined['Last Year Bonus + Stocks'] = df_refined['Last Year Bonus + Stocks'].str.replace(',', '')
df_refined['Last Year Bonus + Stocks'] = pd.to_numeric(df_refined['Last Year Bonus + Stocks'], errors='coerce')

df_refined['Last Year Total Comp'] = df_refined['Bonus + Stocks'] + df_refined['Base Salary']
df_refined.head()
```

**Part C: Analysis**

```python
#Step 1: 
# What are the top 6 used Programming Languages? 

n = 6
df['Programming Language'].value_counts().index.tolist()[:n]





```

```python
#Step 2: Let's see the top 20 used Programming Languages. 


df['Programming Language'].value_counts()[:20].plot(kind='barh', color='brown') 
```

```python
# Step 3.  Let's see the means of the Base Salary by Gender...

df.groupby('Gender')['Base Salary'].mean()
```

```python
# Let's polish the data further -- it turns out we have the variable 'Experience' as string, 
# because some respondents gave wordy answers. Let's fix that. 


df['Experience'] = df['Experience'].str.replace(',','.')

df = df[df.Experience != '1 (as QA Engineer) / 11 in total']


df = df[df.Experience != '15. thereof 8 as CTO']


df = df[df.Experience != '6 (not as a data scientist. but as a lab scientist)']


df = df[df.Experience != 'less than year']
```

```python
# Could the gap between the salaries be because of different experience of the groups?  


df['Experience'] = df['Experience'].astype(float)

df.groupby('Gender')['Experience'].mean()

```

Well, seems so, at least partially.

```python
# Now let's visualize how the experience correlates with the salary! 
# For that we need to get rid of the outliers in both experience and base salary. 

df = df[df['Experience'] < 10] 

df = df[df['Base Salary'] < 100000] 

```

```python
# Let's Plot!

(ggplot(df, aes(x='Experience', y='Base Salary'))+
       geom_point(color='steelblue')+
 theme_bw()
)
```

Well, there is a unequivocal increase in salary with more experience!
